```python
import pandas as pd
import seaborn as se
import numpy as np

import matplotlib.pyplot as plt
import seaborn as se
import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
from keras import callbacks
from sklearn.metrics import confusion_matrix
import pickle
import os
```

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
img_x,img_y=28,28
```

```python
#The input shape that CNN expects is a 4D array (batch, height, width, channels).
#Channels signify whether the image is grayscale or colored. 
#In our case, we are using grayscale images 
#so we give 1 for channels if these are colored images we give 3(RGB)
X_train = X_train.reshape(X_train.shape[0],img_x,img_y,1)
X_test =X_test.reshape(X_test.shape[0],img_x,img_y,1)
input_shape = (img_x,img_y,1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train =X_train.astype('float')
X_test = X_test.astype('float')
```

```python
X_train /=255
X_test /=255
```

```python
X_train[0]
```

```python
X_test.shape
```

```python
# convert class vectors to binary class matrices - this is for use in the
# categorical_crossentropy loss below
```

```python
y_train = keras.utils.to_categorical(y_train,10) #10 digits are there
```

```python
y_test = keras.utils.to_categorical(y_test,10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)
```

```python
classifier = Sequential()
```

```python
classifier.add(Conv2D(filters=32,kernel_size=(5,5),activation='relu',input_shape=(28,28,1)))
```

```python
#classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
classifier.add(Conv2D(filters=64,kernel_size=(5,5),activation='relu'))
```

```python
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
classifier.add(Flatten())
```

```python
#classifier.add(Conv2D(filters=64,kernel_size=(4,4),activation='relu'))
```

```python
#classifier.add(Conv2D(filters=64,kernel_size=(4,4),activation='relu'))
```

```python
classifier.add(Dense(1000,activation='relu'))
```

```python
classifier.add(Dense(10,activation='softmax'))
```

```python
classifier.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
class AccuracyHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.acc = []

    def on_epoch_end(self, batch, logs={}):
        self.acc.append(logs.get('acc'))

history = AccuracyHistory()
```

```python
classifier.fit(X_train,y_train,batch_size=64,epochs=20,validation_data=(X_test,y_test),verbose=1,callbacks=[history])
```

```python
y_pred = classifier.predict(X_test)
```

```python
score = classifier.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
